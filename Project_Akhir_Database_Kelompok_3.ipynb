```python
# Handle missing values
# 1. BuildingArea - Fill with median
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].median())

# 2. YearBuilt - Fill with median
df['YearBuilt'] = df['YearBuilt'].fillna(df['YearBuilt'].median())

# 3. CouncilArea - Fill with mode
df['CouncilArea'] = df['CouncilArea'].fillna(df['CouncilArea'].mode()[0])

# 4. Car - Fill with median
df['Car'] = df['Car'].fillna(df['Car'].median())

# Drop rows with any missing values
df = df.dropna()

# Verify no missing values remain
missing_values_percent = df.isnull().mean() * 100
print(missing_values_percent)
```

```python
# Drop Lattitude and Longtitude attributes
df = df.drop(columns=['Lattitude', 'Longtitude'])
```

```python
def df_ruangan(df):
  no_rooms = df['Rooms'].astype(str) #konversi ke string
  bedrooms = df['Bedroom2'].astype(str).str[0]
  bathrooms = df['Bathroom'].astype(str).str[0]
  car = df['Car'].astype(str).str[0]
  df['rooms_id'] = "R"+ no_rooms + bedrooms + bathrooms + car #menggabungkan string menjadi rooms_id

  return df
```

#Making Database

```python
import sqlite3
conn = sqlite3.connect("Real_estate_database.db")
cur = conn.cursor()
```

```python
try:
    cur.execute("""CREATE TABLE IF NOT EXISTS tabel_ruangan (
        Rooms_id varchar(50) PRIMARY KEY NOT NULL UNIQUE,
        no_rooms INTEGER NOT NULL,
        Bedrooms INTEGER NOT NULL,
        Bathrooms INTEGER NOT NULL,
        Car INTEGER NOT NULL,
        UNIQUE(Rooms_id)
    )""")
except sqlite3.ProgrammingError as e:
    print(f"Error creating tabel_ruangan: {e}")

try:
    cur.execute("""CREATE TABLE IF NOT EXISTS tabel_lokasi (
        Location_id varchar(50) PRIMARY KEY NOT NULL UNIQUE,
        Region_name varchar(50) NOT NULL,
        Council_Area varchar(50) NOT NULL,
        Suburb varchar(50) NOT NULL,
        Postcode INTEGER NOT NULL,
        Property_count INTEGER NOT NULL,
        UNIQUE(Location_id)
    )""")
except sqlite3.ProgrammingError as e:
    print(f"Error creating tabel_lokasi: {e}")

try:
    cur.execute("""CREATE TABLE IF NOT EXISTS tabel_property (
        Property_ID INTEGER PRIMARY KEY NOT NULL UNIQUE,
        Rooms_id varchar(50),
        Location_id varchar(50),
        Type varchar(50) UNIQUE,
        Distance FLOAT(3) NOT NULL,
        Landsize INTEGER NOT NULL,
        Building_Area INTEGER NOT NULL,
        YearBuilt INTEGER NOT NULL,
        Adress varchar(50) NOT NULL,
        UNIQUE(Property_ID),
        FOREIGN KEY (Rooms_id) REFERENCES tabel_ruangan(Rooms_id),
        FOREIGN KEY (Location_id) REFERENCES tabel_lokasi(Location_id)
    )""")
except sqlite3.ProgrammingError as e:
    print(f"Error creating tabel_property: {e}")

try:
    cur.execute("""CREATE TABLE IF NOT EXISTS tabel_agen_penjualan (
        Sale_ID varchar(10) PRIMARY KEY NOT NULL UNIQUE,
        Property_ID INTEGER,
        Date DATE NOT NULL,
        Method varchar(5) NOT NULL,
        Price INTEGER NOT NULL,
        SellerG varchar(10) NOT NULL,
        UNIQUE(Sale_ID),
        FOREIGN KEY (Property_ID) REFERENCES tabel_property(Property_ID)
    )""")
except sqlite3.ProgrammingError as e:
    print(f"Error creating tabel_agen_penjualan: {e}")

# Commit the changes
conn.commit()
```

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("Real_estate_database.db")

# Mengambil lima baris pertama
def fetch_table_head(table_name):
    query = f"SELECT * FROM {table_name} LIMIT 5;"
    df = pd.read_sql_query(query, conn)
    return df

# Specify the table names
table_names = ["tabel_ruangan", "tabel_lokasi", "tabel_property", "tabel_agen_penjualan"]

# Fetch and display head of each table
table_data = {}
for table_name in table_names:
    df = fetch_table_head(table_name)
    table_data[table_name] = df

# Close the connection
conn.close()

# Display tables in pandas DataFrame format
for table_name, df in table_data.items():
    print(f"\nHead of {table_name}:")
    display(pd.DataFrame(df))
```

```python
def simulate(area):
    try:
        # Membaca data dari file CSV
        df = pd.read_csv("melb_data.csv")

        # Cetak semua area yang ada di data untuk pengecekan
        print("Available areas in data:", df['Suburb'].unique())

        # Filter data berdasarkan area
        df_area = df[df['Suburb'].str.lower() == area.lower()]

        if df_area.empty:
            print(f"Tidak ada data untuk area {area}.")
        else:
            # Menghitung harga termurah
            cheapest_price = df_area['Price'].min()
            cheapest_suburb = df_area.loc[df_area['Price'].idxmin()]

            # Menghitung agen dengan penjualan terbanyak
            top_agent = df_area['SellerG'].mode().values[0]

            print(f"Harga termurah di area {area}: ${cheapest_price:.2f}")
            print(f"Agen dengan penjualan terbanyak: {top_agent}")
            print(f"Rata-rata harga rumah di {area}: ${df_area['Price'].mean():.2f}")
            print(f"Rumah termurah di {area}: Lokasi: {cheapest_suburb['Address']}, Room: {cheapest_suburb['Rooms']}")
    except Exception as e:
        print(f"Error: {e}")
```

```python
import sqlite3
from datetime import datetime
import ipywidgets as widgets
from IPython.display import display

# Fungsi untuk memasukkan data ke dalam database
def insert_data(Sale_ID, Property_ID, Date, Method, Price, SellerG):
    conn = sqlite3.connect('Real_estate_database.db')
    cur = conn.cursor()
    cur.execute("INSERT INTO tabel_agen_penjualan (Sale_ID, Property_ID, Date, Method, Price, SellerG) VALUES (?, ?, ?, ?, ?, ?)",
                (Sale_ID, Property_ID, Date, Method, Price, SellerG))
    conn.commit()
    conn.close()

# Fungsi untuk menampilkan data dari database
def view_data():
    conn = sqlite3.connect('Real_estate_database.db')
    cur = conn.cursor()
    cur.execute("SELECT * FROM tabel_agen_penjualan ORDER BY Property_ID DESC LIMIT 10")
    rows = cur.fetchall()
    for row in rows:
        print(row)
    conn.close()

# Definisikan widgetsnya
sale_id_input = widgets.Text(value='', description="Sale ID:", disabled=False, style={'description_width': 'initial'})
property_id_input = widgets.IntText(value=0, description="ID Property:", disabled=False, style={'description_width': 'initial'})
date_input = widgets.DatePicker(description="Date (DD/MM/YYYY):", disabled=False, style={'description_width': 'initial'})
method_input = widgets.Text(value='', description="Method:", disabled=False, style={'description_width': 'initial'})
price_input = widgets.IntText(value=0, description="Price:", disabled=False, style={'description_width': 'initial'})
seller_input = widgets.Text(value='', description="Seller:", disabled=False, style={'description_width': 'initial'})

submit_button = widgets.Button(description="Submit", disabled=False, button_style='success', tooltip="Click to submit data")
view_button = widgets.Button(description="View Data", disabled=False, button_style='info', tooltip="Click to view data")

# mendefinisikan button click nya
def on_submit_button_clicked(b):
    Sale_ID = sale_id_input.value
    Property_ID = property_id_input.value
    Date = date_input.value.strftime('%d/%m/%Y')
    Method = method_input.value
    Price = price_input.value
    SellerG = seller_input.value
    insert_data(Sale_ID, Property_ID, Date, Method, Price, SellerG)
    print("Data inserted successfully!")

def on_view_button_clicked(b):
    view_data()

submit_button.on_click(on_submit_button_clicked)
view_button.on_click(on_view_button_clicked)

# menampilkan widgets
display(sale_id_input, property_id_input, date_input, method_input, price_input, seller_input, submit_button, view_button)
```

```python
import sqlite3
from datetime import datetime
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# Fungsi untuk memasukkan data ke dalam database
def insert_data(Sale_ID, Property_ID, Date, Method, Price, SellerG):
    conn = sqlite3.connect('Real_estate_database.db')
    cur = conn.cursor()
    try:
        cur.execute("INSERT INTO tabel_agen_penjualan (Sale_ID, Property_ID, Date, Method, Price, SellerG) VALUES (?, ?, ?, ?, ?, ?)",
                    (Sale_ID, Property_ID, Date, Method, Price, SellerG))
        conn.commit()
        print("Data inserted successfully!")
    except sqlite3.IntegrityError:
        print(f"Error: Sale_ID '{Sale_ID}' already exists. Please use a different Sale_ID.")
    conn.close()

# Fungsi untuk menampilkan data dari database
def view_data():
    conn = sqlite3.connect('Real_estate_database.db')
    cur = conn.cursor()
    cur.execute("SELECT * FROM tabel_agen_penjualan ORDER BY Property_ID DESC LIMIT 10")
    rows = cur.fetchall()
    conn.close()
    return rows

# Definisikan widgetsnya
sale_id_input = widgets.Text(value='', description="Sale ID:", disabled=False, style={'description_width': 'initial'})
property_id_input = widgets.IntText(value=0, description="ID Property:", disabled=False, style={'description_width': 'initial'})
date_input = widgets.DatePicker(description="Date (DD/MM/YYYY):", disabled=False, style={'description_width': 'initial'})
method_input = widgets.Text(value='', description="Method:", disabled=False, style={'description_width': 'initial'})
price_input = widgets.IntText(value=0, description="Price:", disabled=False, style={'description_width': 'initial'})
seller_input = widgets.Text(value='', description="Seller:", disabled=False, style={'description_width': 'initial'})

submit_button = widgets.Button(description="Submit", disabled=False, button_style='success', tooltip="Click to submit data")
view_button = widgets.Button(description="View Data", disabled=False, button_style='info', tooltip="Click to view data")

# Mendefinisikan button click
def on_submit_button_clicked(b):
    Sale_ID = sale_id_input.value
    Property_ID = property_id_input.value
    Date = date_input.value.strftime('%d/%m/%Y')
    Method = method_input.value
    Price = price_input.value
    SellerG = seller_input.value

    # Check if Sale_ID already exists
    conn = sqlite3.connect('Real_estate_database.db')
    cur = conn.cursor()
    cur.execute("SELECT Sale_ID FROM tabel_agen_penjualan WHERE Sale_ID = ?", (Sale_ID,))
    existing_sale_id = cur.fetchone()
    conn.close()

    clear_output()  # Clear previous output

    if existing_sale_id:
        print(f"Error: Sale_ID '{Sale_ID}' already exists. Please use a different Sale_ID.")
    else:
        insert_data(Sale_ID, Property_ID, Date, Method, Price, SellerG)

    display_widgets()  # Redisplay widgets after clearing output

def on_view_button_clicked(b):
    rows = view_data()
    clear_output()  # Clear previous output
    display_widgets()  # Redisplay widgets

    if rows:
        df = pd.DataFrame(rows, columns=['Sale_ID', 'Property_ID', 'Date', 'Method', 'Price', 'SellerG'])
        display(df)
    else:
        print("No data found.")

def display_widgets():
    display(sale_id_input, property_id_input, date_input, method_input, price_input, seller_input, submit_button, view_button)

submit_button.on_click(on_submit_button_clicked)
view_button.on_click(on_view_button_clicked)

# Menampilkan widgets
display_widgets()
```

```python
import sqlite3
from datetime import datetime
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Connect to the SQLite database to fetch unique SellerG values
conn = sqlite3.connect('Real_estate_database.db')
cur = conn.cursor()
cur.execute("SELECT DISTINCT SellerG FROM tabel_agen_penjualan")
seller_options = [row[0] for row in cur.fetchall()]
conn.close()

# Function to generate Sale_ID
def generate_sale_id(method, property_id, seller):
    seller_code = seller[:2].upper()  # Simplified example of generating Seller_Code
    return f"{method}{property_id}{seller_code}"

# Function to insert data into the database
def insert_data(Sale_ID, Property_ID, Date, Method, Price, SellerG):
    conn = sqlite3.connect('Real_estate_database.db')
    cur = conn.cursor()
    try:
        cur.execute("INSERT INTO tabel_agen_penjualan (Sale_ID, Property_ID, Date, Method, Price, SellerG) VALUES (?, ?, ?, ?, ?, ?)",
                    (Sale_ID, Property_ID, Date, Method, Price, SellerG))
        conn.commit()
        print("Data inserted successfully!")
    except sqlite3.IntegrityError:
        print(f"Error: Sale_ID '{Sale_ID}' already exists. Please use a different Sale_ID.")
    conn.close()

# Function to view data from the database
def view_data():
    conn = sqlite3.connect('Real_estate_database.db')
    cur = conn.cursor()
    cur.execute("SELECT * FROM tabel_agen_penjualan ORDER BY Property_ID DESC LIMIT 10")
    rows = cur.fetchall()
    conn.close()
    return pd.DataFrame(rows, columns=['Sale_ID', 'Property_ID', 'Date', 'Method', 'Price', 'SellerG'])

# Define widgets
sale_id_input = widgets.Text(value='', description="Sale ID:", disabled=True, style={'description_width': 'initial'})
property_id_input = widgets.IntText(value=0, description="ID Property:", disabled=False, style={'description_width': 'initial'})
date_input = widgets.DatePicker(description="Date (YYYY-MM-DD):", disabled=False, style={'description_width': 'initial'})
method_input = widgets.Text(value='', description="Method:", disabled=False, style={'description_width': 'initial'})
price_input = widgets.IntText(value=0, description="Price:", disabled=False, style={'description_width': 'initial'})
seller_input = widgets.Dropdown(options=seller_options, description="Seller:", disabled=False, style={'description_width': 'initial'})

submit_button = widgets.Button(description="Submit", disabled=False, button_style='success', tooltip="Click to submit data")
view_button = widgets.Button(description="View Data", disabled=False, button_style='info', tooltip="Click to view data")

# Define button click handlers
def on_submit_button_clicked(b):
    Sale_ID = generate_sale_id(method_input.value, property_id_input.value, seller_input.value)
    Property_ID = property_id_input.value
    Date = date_input.value.strftime('%Y-%m-%d')
    Method = method_input.value
    Price = price_input.value
    SellerG = seller_input.value

    insert_data(Sale_ID, Property_ID, Date, Method, Price, SellerG)
    sale_id_input.value = Sale_ID  # Display the generated Sale_ID

def on_view_button_clicked(b):
    df = view_data()
    print(df)

submit_button.on_click(on_submit_button_clicked)
view_button.on_click(on_view_button_clicked)

# Display widgets
display(sale_id_input, property_id_input, date_input, method_input, price_input, seller_input, submit_button, view_button)
```

#Case 3
Misalkan terdapat user yang ingin mencari informasi mengenai properti di Melbourne yang sudah dijual dengan pertimbangan budget yang ia miliki dan juga tipe properti yang akan ia cari. User ingin mendapatkan informasi berdasarkan jarak terdekat dengan pusat kota karena ia memiliki tempat kerja di Ibu Kota.

```python
import sqlite3
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Connect to the SQLite database to fetch unique property types
conn = sqlite3.connect('Real_estate_database.db')
cur = conn.cursor()
cur.execute("SELECT DISTINCT Type FROM tabel_property")
type_options = [row[0] for row in cur.fetchall()]
conn.close()

# Define widgets
price_range_input = widgets.FloatRangeSlider(
    value=[85000, 9000000],
    min=85000,
    max=9000000,
    step=10000,
    description='Price Range:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.0f',
    style={'description_width': 'initial'}
)
type_input = widgets.Dropdown(
    options=type_options,
    description='Type:',
    disabled=False,
    style={'description_width': 'initial'}
)
search_button = widgets.Button(description="Search", disabled=False, button_style='success', tooltip="Click to search properties")
output_df = widgets.Output()

# Function to fetch and display data
def search_properties(price_range, property_type):
    conn = sqlite3.connect('Real_estate_database.db')
    query = """
    SELECT p.Property_ID, p.rooms_id, r.Rooms, r.Bedroom2, r.Bathroom, r.Car,
           p.Type, p.Distance, p.Landsize, p.BuildingArea, p.YearBuilt, p.Address,
           s.Price, s.Date, s.Method, s.SellerG
    FROM tabel_property p
    JOIN tabel_ruangan r ON p.rooms_id = r.rooms_id
    JOIN tabel_agen_penjualan s ON p.Property_ID = s.Property_ID
    WHERE s.Price BETWEEN ? AND ? AND p.Type = ?
    ORDER BY p.Distance ASC
    LIMIT 10;
    """
    df = pd.read_sql_query(query, conn, params=(price_range[0], price_range[1], property_type))
    conn.close()
    return df

# Define button click handler
def on_search_button_clicked(b):
    with output_df:
        output_df.clear_output()
        price_range = price_range_input.value
        property_type = type_input.value
        df = search_properties(price_range, property_type)
        display(df)

search_button.on_click(on_search_button_clicked)

# Display widgets
initial_message = widgets.Label(value="Keterangan:\n h: rumah,cottage,villa,semi-terrace;\n u: unit,duplex;\n t: townhouse.")
display(price_range_input, initial_message, type_input, search_button, output_df)
```
